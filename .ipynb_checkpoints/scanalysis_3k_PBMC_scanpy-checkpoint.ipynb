# Preprocessing AND Quality Control

```python
# Getting How many Genes are there that are expressed in less than 3 cells 
print(f"Min Number of cells a gene is expressed  : {min(adata.var["n_cells_by_counts"])} ")
print(f"Max Number of cells a gene is expressed : {max(adata.var["n_cells_by_counts"])} ")
Gene_exp = adata.var_names[(adata.var["n_cells_by_counts"] < 3 )].tolist()
print(f"Genes Expressed in less than 3 cells : {len(Gene_exp)} \n ")  

# Getting How many Cells have less than 100 genes expressed 
print(f"Min Number of genes in a cell : {min(adata.obs["n_genes_by_counts"])} ")
print(f"Max Number of genes in a cell : {max(adata.obs["n_genes_by_counts"])} ")
Per_cell_gene_exp = adata.obs_names[(adata.obs["n_genes_by_counts"]< 200 )].tolist()
print(f"Number of cells with less than 200 genes expressed ( Insignificant ) : {len(Per_cell_gene_exp)}")
```

```python
# BASIC FILTERING 
sc.pp.filter_genes(adata, min_cells = 3 ) 
print(f"Removed {len(Gene_exp)} Genes")

sc.pp.filter_cells(adata, min_genes = 200 )
print(f"Removed {len(Per_cell_gene_exp)} cells ")


Gene_exp = adata.var_names[(adata.var["n_cells_by_counts"] < 3)].tolist()
Per_cell_gene_exp = adata.obs_names[(adata.obs["n_genes_by_counts"]< 200 )].tolist()
# print(f"Genes Expressed in less than 3 cells : {len(Gene_exp)} ")
# print(f"Cells with less than 200 gene expression : {len(Per_cell_gene_exp)} ")
```

```python
# Calculating the QC Metrices Of MT, RIBO and HEMO CONTENt
sc.pp.calculate_qc_metrics(adata_qc, log1p=False,percent_top=None, qc_vars=["mt","ribo","hemoglobin"], inplace=True)
```

```python
# More Structured Through AI

# Plotting qc metrices

# Set up the figure and axes objects for a 3x2 grid
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 18))
# axes is now a 2D numpy array: [[ax1, ax2], [ax3, ax4], [ax5, ax6]]

# --- Row 1: Genes per Cell ---
# 1a: Histogram
ax1_hist = axes[0, 0]
ax1_hist.hist(adata_qc.obs["n_genes_by_counts"], bins=50, alpha=0.7, color="steelblue", edgecolor='black')
ax1_hist.set_xlabel("Genes per cell")
ax1_hist.set_ylabel("Number of cells")
ax1_hist.set_title("Gene distribution (Histogram)")
median_genes = np.median(adata.obs["n_genes_by_counts"])
ax1_hist.axvline(median_genes, color="red", linestyle="--", label=f"Median {median_genes:.0f}")
ax1_hist.legend()

# 1b: Violin Plot
ax1_viol = axes[0, 1]
sc.pl.violin(adata_qc, "n_genes_by_counts", jitter=0.4, ax=ax1_viol, show=False )
ax1_viol.set_title("Genes per cell (Violin Plot)")
ax1_viol.set_xlabel("") # Clear redundant x-label

# --- Row 2: UMI Counts per Cell ---
# 2a: Histogram
ax2_hist = axes[1, 0]
ax2_hist.hist(adata_qc.obs["total_counts"], bins=50, alpha=0.7, color="yellow", edgecolor='black')
ax2_hist.set_xlabel("UMI Detected per cell")
ax2_hist.set_ylabel("Number of cells")
ax2_hist.set_title("UMI distribution (Histogram)")
median_counts = np.median(adata.obs["total_counts"])
ax2_hist.axvline(median_counts, color="red", linestyle="--", label=f"Median {median_counts:.0f}")
ax2_hist.legend()

# 2b: Violin Plot
ax2_viol = axes[1, 1]
sc.pl.violin(adata_qc, "total_counts", jitter=0.4, ax=ax2_viol, show=False)
ax2_viol.set_title("UMI Counts per cell (Violin Plot)")
ax2_viol.set_xlabel("")

# --- Row 3: Mitochondrial Distribution by Percentage---

# 3a: Histogram
ax3_hist = axes[2, 0]
ax3_hist.hist(adata_qc.obs["pct_counts_mt"], bins=50, alpha=0.7, color="coral", edgecolor='black')
ax3_hist.set_xlabel("% MT counts per cell")
ax3_hist.set_ylabel("Number of cells")
ax3_hist.set_title("Mitochondrial % distribution (Hist)")
median_mt = np.median(adata_qc.obs["pct_counts_mt"])
ax3_hist.axvline(median_mt, color="red", linestyle="--", label=f"Median {median_mt:.2f}%")
ax3_hist.legend()

# 3b: Violin Plot
ax3_viol = axes[2, 1]
sc.pl.violin(adata_qc, "pct_counts_mt", jitter=0.4, ax=ax3_viol, show=False)
ax3_viol.set_title("Mitochondrial % (Violin Plot)")
ax3_viol.set_xlabel("")


# Adjust layout to prevent titles/labels from overlapping
plt.tight_layout()
plt.show()
```

```python
sc.pp.highly_variable_genes(
    adata_qc,
    n_top_genes=2000,
    min_mean=0.0125,
    max_mean=3,
    min_disp=0.5,
    flavor="seurat_v3",
    layer="counts",
    inplace=True
)
```

```python
sc.pl.highly_variable_genes(adata_qc)
```

```python
# Seclecting only highly variable genes out of 13714 genes 
adata_qc = adata_qc[:, adata_qc.var.highly_variable]
```

## Dimensionality Reduction 

```python
# Calculating PCs 
sc.tl.pca(adata_qc)

# BY DEFAULT IT CALCULATES 50 PCs
# Results stored in adata_qc.obsm["X_pca"]
```

```python
# Visualizing PCAs

sc.pl.pca_variance_ratio(adata_qc, n_pcs=50, log=True)
```

```python
sc.pl.pca(
    adata_qc,
    color=[ "pct_counts_mt"],
    dimensions=[ (0, 1), (2, 3)],
    annotate_var_explained=True,
)
```

```python
adata_qc.layers["scaled"] = adata_qc.X.toarray()
sc.pp.regress_out(adata_qc, ["total_counts", "pct_counts_mt"], layer="scaled")
# RemovesIt removes the linear effects of specified variables Total_counts and pct_counts_mt's data from scaled layer
sc.pp.scale(adata_qc, max_value=10, layer="scaled")
# This applies z-score normalization (scaling) to the data within the "scaled" layer.
```

```python
sc.tl.pca(adata_qc)
```

```python
sc.pl.pca(
    adata_qc,
    color=[ "pct_counts_mt"],
    dimensions=[ (0, 1), (2, 3)],
    annotate_var_explained=True,
)
```

## CELL ANNOTATION - MARKER GENE IDENTIFICATION

This data will require to find the marker genes Manually ,  based on each gene's expression in a cell cluster , if in a cell cluster a gene is highly expressed in comparison to other clusters or cells then possibly it's a marker gene.

For Example :  We'll test The GENE A's Expression in Two groups 

Group 1 = cells of our desired cluster

Group 2 = All other cells 

If the GENE A's expression in GROUP 1 is higher than GROUP 2 Then it'll be a marker gene of the Cells in Group 1

The same test is done for every gene and for every cluster.

Result: each cluster gets a ranked list of genes

Basically Top genes = markers
